### Обработка этажей вручную

```python
data = pd.read_csv(zip_file.open(zip_file.namelist()[0]), low_memory=False)
data
```

```python
floor(data)
```

```python
is_nan_list = data['floor'].isnull()

for i in tqdm(range(len(data))):
    if not is_nan_list[i]:
        try:
            data['floor'][i] = float(str(data['floor'][i]))
        except:
            try:
                data['floor'][i] = float(reg_float(data['floor'][i]))
            except:
                data['floor'][i] = float(0)
```

```python
data.to_csv("../data/test.zip", 
           index=False, 
           compression="zip")
```

### Обработка этажей категориально

```python
data = pd.read_csv(zip_file.open(zip_file.namelist()[0]), low_memory=False)
floor(data)

is_nan_list = data['floor'].isnull()

for i in tqdm(range(len(data))):
    try:
        data['floor'][i] = float(str(data['floor'][i]))
    except:
        try:
            data['floor'][i] = float(reg_float(data['floor'][i]))
        except:
            data['floor'][i] = float(0)
```

```python
for i in tqdm(range(len(data))):
    if not is_nan_list[i]:
        if data['floor'][i] <= 0:
            data['floor'][i] = 'basement'
        elif data['floor'][i] == 1:
            data['floor'][i] = 'first'
        else:
            data['floor'][i] = 'multy_level'
    else:
        data['floor'][i] = 'unknown'
```

```python
data.to_csv("../data/mult.zip", 
           index=False, 
           compression="zip")
```
