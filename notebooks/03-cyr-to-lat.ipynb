```python
cols_to_rename = {}
for column in poi_cols:
    try:
        cols_to_rename[column] = translit(column, reversed=True).replace(' ', '_').replace("'", "").replace('&', '')
    except:
        cols_to_rename[column] = column.replace(' ', '_')
```
