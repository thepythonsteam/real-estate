### Анализ стоимсоти недвижимости по городом за все время

```python
import pandas as pd
from tqdm import tqdm
import zipfile
from prophet import Prophet
```

### Данные:

```python
path_to_train = '../data/train.csv.zip'
zip_file = zipfile.ZipFile(path_to_train)
train_df = pd.read_csv(zip_file.open(zip_file.namelist()[0]))
```

```python
# города
cities_list = list(train_df['city'])
cities = set(cities_list)

cities_counts = []
for city in tqdm(cities):
    cities_counts.append(cities_list.count(city))

print(min(cities_counts), max(cities_counts))
```

```python
# даты 
dates_list = list(train_df['date'])
dates = sorted(set(dates_list))

dates_counts = []
for date in tqdm(dates):
    dates_counts.append(dates_list.count(date))

print(dates_counts)
# print(sorted(set(train_df['date'])))
```
